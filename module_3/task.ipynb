## 2) Убедитесь, что номера заказов уникальны и не дублируются по строкам

```python
# Проверка на дубликаты
if (
    orders["order_id"].duplicated().sum()
    == orders_short["order_id"].duplicated().sum()
    == 0
):
    print("Дубликаты не найдены")
else:
    print('Дубликаты "orders":', orders["order_id"].duplicated().sum())
    print('Дубликаты "orders_short":', orders_short["order_id"].duplicated().sum())

    print("Обнаружены дубликаты, исправление...")

    # Если дубликаты есть - убираем их
    orders = orders.drop_duplicates(subset=["order_id"])
    orders_short = orders_short.drop_duplicates(subset=["order_id"])

    print('Дубликаты "orders":', orders["order_id"].duplicated().sum())
    print('Дубликаты "orders_short":', orders_short["order_id"].duplicated().sum())
```

## 3) Проверьте наличие требуемых колонок: store_id, order_price, profit, delivery_distance, planned_prep_time, region_id, items_count, prepared_on_time

```python
# Список требуемых столбцов
REQUIRED_COLUMNS: list[str] = [
    "store_id",
    "order_price",
    "profit",
    "delivery_distance",
    "planned_prep_time",
    "region_id",
    "items_count",
    "prepared_on_time",
]


def check_dataframe(df: pd.DataFrame) -> bool:
    """
    Проверяет наличие всех требуемых столбцов в DataFrame.
    """
    for column in REQUIRED_COLUMNS:
        if column not in df.columns:
            return False
    return True


orders_short_check = check_dataframe(orders_short)
orders_check = check_dataframe(orders)

if orders_short_check or orders_check:
    print("Все требуемые столбцы присутствуют")
else:
    raise Exception("Некоторые столбцы отсутствуют")
```

## 1) Разделите данные на признаки (X) и целевую переменную (у), где prepared_on_time — цель

```python
# Получение целевых значений
y = orders["prepared_on_time"]
y.head()
```

## 1) Постройте модель логистической регрессии с учётом class_weight='balanced'

```python
from sklearn.linear_model import LogisticRegression

# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
model = LogisticRegression(class_weight="balanced", max_iter=500)
```

## 2) Подберите оптимальные гиперпараметры на кросс-валидации (GridSearchCV или аналог)

```python
from sklearn.experimental import enable_halving_search_cv  # noqa: F401
from sklearn.model_selection import HalvingRandomSearchCV


def tune_model(
    model: LogisticRegression, params: dict, x_train, y_train
) -> HalvingRandomSearchCV:
    # https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-string-names
    SCORING: str = "f1"

    # https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.HalvingRandomSearchCV.html
    search_cv = HalvingRandomSearchCV(model, params, cv=5, scoring=SCORING)

    search_cv.fit(x_train, y_train)

    return search_cv
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
params = {
    "C": [0.01, 0.1, 1.0, 5.0, 10.0, 50.0, 100.0],
    "penalty": ["l1", "l2"],
    "solver": ["liblinear"],
    "max_iter": [100, 200, 500],
}

search_cv = tune_model(model, params, X_train, y_train)

print("Лучшие гиперпараметры:", search_cv.best_params_)
print("Лучшая средняя метрика CV:", search_cv.best_score_)

best_model = search_cv.best_estimator_
```

## 4) Обучите модель с найденными параметрами на всей обучающей выборке

```python
result = best_model.fit(X_train_scaled, y_train)
result
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
)


y_pred_train = best_model.predict(X_train_scaled)
y_pred_train

train_scores = {}

# https://sky.pro/wiki/analytics/chto-takoe-metrika-accuracy-opredelenie-osobennosti-primenenie
train_scores["accuracy"] = accuracy_score(y_train, y_pred_train)

# https://sky.pro/wiki/analytics/precision-i-recall-chto-eto-takoe-i-kak-vliyayut-na-rezultaty
train_scores["precision"] = precision_score(y_train, y_pred_train)
train_scores["recall"] = recall_score(y_train, y_pred_train)

# https://sky.pro/wiki/analytics/f1-score-v-sklearnmetrics-otsenka-tochnosti-modelej-mashinnogo-obucheniya
train_scores["f1-score"] = f1_score(y_train, y_pred_train)

# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html
train_scores["classification"] = classification_report(y_train, y_pred_train)

# https://sky.pro/wiki/analytics/chto-takoe-confusion-matrix--metrika-otsenki-kachestva-klassifikatsii
train_scores["confusion"] = confusion_matrix(y_train, y_pred_train)

for k, v in train_scores.items():
    print(f"{k}:\n{v}\n")
```
